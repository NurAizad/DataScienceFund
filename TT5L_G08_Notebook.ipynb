```python
#FINDING OUTLIERS
def findingOutliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1

    lowerLimit = Q1 - 1.5*IQR
    upperLimit = Q3 + 1.5*IQR
    return (column < lowerLimit) | (column > upperLimit)
```

```python
numerical_column = [ "num_units", "injuries_total", "injuries_fatal", "injuries_incapacitating",
                     "injuries_non_incapacitating", "injuries_reported_not_evident", "injuries_no_indication" ]
```

### Clustering

```python
accidents_cluster = accidents_clean.copy()
```

```python
accidents_cluster.head()
```

```python
cols = ['weather_condition', 'lighting_condition', 'roadway_surface_cond', 'road_defect', 'alignment']
accidents_cluster[cols].nunique()
```

```python
accidents_cluster = pd.get_dummies(accidents_cluster, columns = cols, dtype= "int") 
#https://www.geeksforgeeks.org/pandas/python-pandas-get_dummies-method/
```

```python
cols_to_check = ['traffic_control_device', 'first_crash_type', 'trafficway_type', 
                 'crash_type', 'intersection_related_i', 'prim_contributory_cause', 'most_severe_injury' ]
accidents_cluster[cols_to_check].nunique()
```

```python
cols_drop = ['traffic_control_device', 'first_crash_type', 'trafficway_type', 'prim_contributory_cause']

accidents_cluster = accidents_cluster.drop(columns = cols_drop)
```

```python
accidents_cluster["damage"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder #https://www.geeksforgeeks.org/machine-learning/how-to-perform-ordinal-encoding-using-sklearn/
import pandas as pd
```

```python
encoder = OrdinalEncoder(categories=[['$500 OR LESS', '$501 - $1,500', 'OVER $1,500']])
accidents_cluster['damage_encoded'] = encoder.fit_transform(accidents_cluster[['damage']])
accidents_cluster = accidents_cluster.drop(columns=['damage'])
```

```python
accidents_cluster['most_severe_injury'].unique()
```

```python
encoder = OrdinalEncoder(categories=[['NO INDICATION OF INJURY', 'REPORTED, NOT EVIDENT',
                                      'NONINCAPACITATING INJURY', 'INCAPACITATING INJURY', 'FATAL' ]])
accidents_cluster['most_severe_injury_encoded'] = encoder.fit_transform(accidents_cluster[['most_severe_injury']])
accidents_cluster = accidents_cluster.drop(columns=['most_severe_injury'])
```

```python
#https://stackoverflow.com/questions/51672709/converting-no-and-yes-into-0-and-1-in-pandas-dataframe
accidents_cluster['intersection_related_i'] = accidents_cluster['intersection_related_i'].map({'Y' : 1, 'N' : 0})
```

```python
accidents_cluster['crash_type'].unique()
```

```python
accidents_cluster['crash_type'] = accidents_cluster['crash_type'].map({'NO INJURY / DRIVE AWAY' : 0,
                                                                       'INJURY AND / OR TOW DUE TO CRASH' : 1})
```

```python
#https://www.geeksforgeeks.org/pandas/pandas-series-dt-dayofweek
#https://www.geeksforgeeks.org/pandas/pandas-series-dt-hour/
accidents_cluster['day_of_week'] = accidents_cluster['crash_date'].dt.dayofweek
accidents_cluster['crash_hour'] = accidents_cluster['crash_date'].dt.hour

accidents_cluster = accidents_cluster.drop(columns = ['crash_date'])
```

```python
accidents_cluster.head()
```

#### Elbow method for Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
ss = StandardScaler()
X = ss.fit_transform(accidents_cluster)
print(accidents_cluster)
print(X)
```

```python
#https://www.w3schools.com/python/python_ml_k-means.asp
inertia =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(accidents_cluster)
    inertia.append(kmeans.inertia_)

plt.plot(range(1,11), inertia, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
model = KMeans(3, verbose=0, n_init='auto')
model.fit(X)
```

```python
accidents_cluster['cluster'] = model.labels_
```

```python
v1 = accidents_cluster['crash_hour']
v2 = accidents_cluster['injuries_total']

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
scatter = ax.scatter(v1, v2, c=accidents_cluster['cluster'],s=50,cmap='jet',alpha=0.7)
ax.set_title('K-Means Clustering')
ax.set_xlabel(v1.name)
ax.set_ylabel(v2.name)
plt.colorbar(scatter)
plt.show()
```
